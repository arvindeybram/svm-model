```python
import re
urllength = []
nodots = []
numberofdigits = []
specialsymbols = []

for i in data.URL:
    numbers = 0
    
    urllength.append(len(i))
    
    dotcount=i.count(".")
    nodots.append(dotcount)
    
    numbers = sum(c.isdigit() for c in i)
    numberofdigits.append(numbers)
    
    new = re.sub('[\w]+' ,'', i)
    specialsymbols.append(len(new)-dotcount)
```

```python
data['SpecialSymbols'] = specialsymbols
data['LengthofUrl'] = urllength
data['DigitsInUrl'] = numberofdigits
data['DotsInUrl'] = nodots
```

```python
from sklearn import svm
from sklearn.cross_validation import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix 
from sklearn.svm import SVC 
```

```python
X_train, X_test, y_train, y_test = train_test_split( X, y, test_size = 0.20, random_state = 100)
```

```python
svm_model_linear = SVC(kernel = 'linear', C = 1).fit(X_train, y_train) 
svm_predictions = svm_model_linear.predict(X_test) 
```

```python
accuracy = svm_model_linear.score(X_test, y_test) 
```
